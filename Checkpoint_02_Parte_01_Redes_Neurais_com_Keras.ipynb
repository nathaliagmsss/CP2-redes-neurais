Importando DataSet

```python
import pandas as pd
```

```python
dt_cal_housing = pd.read_csv('cal_housing.data', low_memory=False, sep=",")
display(dt_cal_housing);
```

# Task
Treinar uma rede neural em Keras para prever o valor médio das casas utilizando o dataset "dt_cal_housing". A rede neural deve ter a seguinte configuração mínima: 3 camadas ocultas com 64, 32 e 16 neurônios, função de ativação ReLU, e uma camada de saída com 1 neurônio e função de ativação Linear. A função de perda a ser utilizada é o Erro Quadrático Médio (MSE).

## Preparação dos dados

### Subtask:
Separar os dados em recursos (features) e rótulos (labels), e dividir o conjunto de dados em conjuntos de treino e teste.


**Reasoning**:
Create the features and labels dataframes and split the data into training and testing sets.



```python
features = dt_cal_housing.iloc[:, :-1]
labels = dt_cal_housing.iloc[:, -1]

from sklearn.model_selection import train_test_split

features_train, features_test, labels_train, labels_test = train_test_split(features, labels, test_size=0.2, random_state=42)

display(features_train.head())
display(features_test.head())
display(labels_train.head())
display(labels_test.head())
```

## Normalização dos dados

Normalizar os dados de entrada para melhorar o desempenho da rede neural.


```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
features_train_scaled = scaler.fit_transform(features_train)
features_test_scaled = scaler.transform(features_test)

display(features_train_scaled[:5])
display(features_test_scaled[:5])
```

## Construção do modelo

### Subtask:
Definir a arquitetura da rede neural em Keras, incluindo as camadas ocultas e a camada de saída com as especificações fornecidas.


**Reasoning**:
Define the neural network architecture as specified in the instructions using Keras Sequential model and Dense layers.



```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(64, activation='relu', input_shape=(features_train_scaled.shape[1],)),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1, activation='linear')
])

model.summary()
```

## Compilação do modelo

Compilar o modelo, especificando a função de perda (MSE) e o otimizador.


**Reasoning**:
Compile the sequential model using the specified loss function (MSE) and an optimizer.



```python
from tensorflow.keras.optimizers import Adam

model.compile(optimizer=Adam(), loss='mse', metrics=['mae'])
```

## Treinamento do modelo

Treinar o modelo com os dados de treino.


**Reasoning**:
Train the compiled Keras model using the training data and validate on the test data as specified in the instructions.



```python
history = model.fit(features_train_scaled, labels_train, epochs=100, batch_size=32, validation_data=(features_test_scaled, labels_test))
```

## Resumo:

### Principais descobertas da análise de dados

* O modelo Keras alcançou uma Perda de Teste (MSE) de aproximadamente 3.94e+09 e um MAE de Teste de aproximadamente 43790.79 nos dados de teste.
* O modelo RandomForest Regressor alcançou um MSE de Teste de aproximadamente 2.506.628.055,13 e um MAE de Teste de aproximadamente 32.041,74 nos dados de teste.
* Com base nas métricas MSE e MAE, o modelo RandomForest Regressor teve melhor desempenho do que o modelo Keras para esta tarefa e conjunto de dados específicos.

### Insights ou próximos passos

* O ajuste adicional de hiperparâmetros para o modelo Keras pode potencialmente melhorar seu desempenho para ser mais competitivo com o RandomForest Regressor.
* Explorar outros algoritmos de regressão do scikit-learn ou diferentes arquiteturas de redes neurais em Keras pode ser benéfico.

## Avaliação do modelo keras

Avaliar o desempenho do modelo nos dados de teste.


**Reasoning**:
Evaluate the trained Keras model using the test data to assess its performance.



```python
loss, mae = model.evaluate(features_test_scaled, labels_test)
print(f'Test Loss (MSE): {loss}')
print(f'Test MAE: {mae}')
```

## Treinamento e avaliação do modelo scikit-learn

### Subtask:
Treinar e avaliar um modelo de regressão do scikit-learn (LinearRegression ou RandomForestRegressor) nos mesmos dados.


```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(features_train_scaled, labels_train)

labels_pred_rf = rf_model.predict(features_test_scaled)

mse_rf = mean_squared_error(labels_test, labels_pred_rf)
mae_rf = mean_absolute_error(labels_test, labels_pred_rf)

print(f'Random Forest Regressor Test MSE: {mse_rf}')
print(f'Random Forest Regressor Test MAE: {mae_rf}')
```

## Comparação dos resultados

Comparar as métricas de desempenho dos modelos Keras e scikit-learn.


**Reasoning**:
Compare the MSE and MAE values from the Keras model and the RandomForestRegressor model to determine which performed better.



```python
print("Comparison of Model Performance:")
print(f"Keras Model Test MSE: {loss}")
print(f"Random Forest Regressor Test MSE: {mse_rf}")
print(f"Keras Model Test MAE: {mae}")
print(f"Random Forest Regressor Test MAE: {mae_rf}")

if mse_rf < loss:
    print("Random Forest Regressor has a lower MSE and performed better.")
elif loss < mse_rf:
    print("Keras Model has a lower MSE and performed better.")
else:
    print("Both models have similar MSE.")

if mae_rf < mae:
    print("Random Forest Regressor has a lower MAE and performed better.")
elif mae < mae_rf:
    print("Keras Model has a lower MAE and performed better.")
else:
    print("Both models have similar MAE.")
```
